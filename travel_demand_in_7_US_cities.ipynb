```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Set Arial as the default font
mpl.rcParams['font.family'] = 'Arial'

# Step 1: Read the data for working days and non-working days
working_days_data = pd.read_csv('C:/Users/justi/Downloads/7cities/7cities/shared bikes raw data/data_visualization/boston_trip_weekday.csv')
non_working_days_data = pd.read_csv('C:/Users/justi/Downloads/7cities/7cities/shared bikes raw data/data_visualization/boston_trip_weekend.csv')

# Step 2: Filter the data for May 2019
working_days_data['date'] = pd.to_datetime(working_days_data['date'], format='%d/%m/%Y', errors='coerce')
non_working_days_data['date'] = pd.to_datetime(non_working_days_data['date'], format='%d/%m/%Y', errors='coerce')

working_days_data = working_days_data[working_days_data['date'].dt.month == 5]
non_working_days_data = non_working_days_data[non_working_days_data['date'].dt.month == 5]

# Step 3: Preprocess the data for working days and non-working days
working_days_data['starttime'] = pd.to_datetime(working_days_data['starttime'], format='%H:%M:%S', errors='coerce')
working_days_data['stoptime'] = pd.to_datetime(working_days_data['stoptime'], format='%M:%S.%f', errors='coerce')

non_working_days_data['starttime'] = pd.to_datetime(non_working_days_data['starttime'], format='%H:%M:%S', errors='coerce')
non_working_days_data['stoptime'] = pd.to_datetime(non_working_days_data['stoptime'], format='%M:%S.%f', errors='coerce')

working_days_data['start_hour'] = working_days_data['starttime'].dt.hour
non_working_days_data['start_hour'] = non_working_days_data['starttime'].dt.hour

# Step 4: Calculate the average travel demand for each hour on working days and non-working days
working_days_demand = working_days_data.groupby('start_hour')['bikeid'].count() / working_days_data['date'].nunique()
non_working_days_demand = non_working_days_data.groupby('start_hour')['bikeid'].count() / non_working_days_data['date'].nunique()

# Step 5: Calculate the overall average travel demand
overall_demand = (working_days_demand + non_working_days_demand) / 2

# Step 6: Create the plot
plt.plot(working_days_demand.index, working_days_demand, 'o-', color='#3776ab', label='Working Days', linewidth=2, markersize=8)
plt.plot(non_working_days_demand.index, non_working_days_demand, 'o-', color='crimson', label='Non-working Days', linewidth=2, markersize=8)
#plt.plot(overall_demand.index, overall_demand, 'o-', color='green', label='Overall', linewidth=2, markersize=8)

# Step 7: Set the x-axis and y-axis labels
plt.xlabel('Time of Day', fontsize= 16)
plt.ylabel('Average Number of Bike Trips',fontsize= 16)

# Step 8: Set the title and legend
plt.title(' Boston (BOS)',fontsize= 16, fontweight='bold')
plt.legend(loc='upper left', fontsize=11)

# Step 9: Set the x-axis ticks and limits
plt.xticks(range(24))
plt.yticks(range(0, 1101, 100))
plt.xlim(0, 23)

# Step 10: Display the plot
plt.tight_layout()
plt.savefig('F5_bos.png', dpi=300, bbox_inches = 'tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Set Arial as the default font
mpl.rcParams['font.family'] = 'Arial'

# Step 1: Read the data for working days and non-working days
data = pd.read_csv('C:/Users/justi/Downloads/7cities/7cities/shared bikes raw data/data_visualization/201905-baywheels-tripdata.csv')

# Step 2: Convert starttime to datetime format
data['starttime'] = pd.to_datetime(data['starttime'], format='%d/%m/%Y %H:%M')

# Step 3: Filter the data for May 2019
data = data[data['starttime'].dt.month == 5]

# Step 4: Preprocess the data for working days and non-working days
data['start_hour'] = data['starttime'].dt.hour
data['day_of_week'] = data['starttime'].dt.dayofweek  # Monday: 0, Sunday: 6

# Exclude specific dates
excluded_dates = ['15/5/2019', '16/5/2019', '18/5/2019', '19/5/2019', '20/5/2019', '21/5/2019', '26/5/2019']
data = data[~data['starttime'].dt.date.isin(pd.to_datetime(excluded_dates, format='%d/%m/%Y').date)]

# Filter data for working days
working_days_data = data[data['day_of_week'] < 5]

# Filter data for non-working days (Saturday, Sunday, and Memorial Day)
non_working_days_data = data[data['day_of_week'].isin([5, 6, 27])]

# Step 5: Calculate the average travel demand for each hour on working days and non-working days
total_working_days = working_days_data['starttime'].dt.date.nunique()
total_non_working_days = non_working_days_data['starttime'].dt.date.nunique()

working_days_demand = working_days_data.groupby('start_hour')['bike_id'].count() / total_working_days
non_working_days_demand = non_working_days_data.groupby('start_hour')['bike_id'].count() / total_non_working_days

# Step 6: Create the plot
plt.plot(working_days_demand.index, working_days_demand, 'o-', color='#3776ab', label='Working Days', linewidth=2, markersize=8)
plt.plot(non_working_days_demand.index, non_working_days_demand, 'o-', color='crimson', label='Non-working Days', linewidth=2, markersize=8)

# Step 7: Set the x-axis and y-axis labels
plt.xlabel('Time of Day', fontsize=16)
plt.ylabel('Average Number of Bike Trips', fontsize=16)

# Step 8: Set the title and legend
plt.title('San Francisco (SFO)', fontsize=16, fontweight='bold')
plt.legend(loc='upper left', fontsize=10)

# Step 9: Set the x-axis ticks and limits
plt.xticks(range(24))
plt.yticks(range(0, 901, 100))
plt.xlim(0, 23)

# Step 10: Display the plot
plt.tight_layout()
plt.savefig('F5_sfo.png', dpi=300, bbox_inches='tight')
plt.show()
```
